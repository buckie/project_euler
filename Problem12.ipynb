The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1
 
 3: 1,3
 
 6: 1,2,3,6

10: 1,2,5,10

15: 1,3,5,15

21: 1,3,7,21

28: 1,2,4,7,14,28
    
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
def tri_num(n):
    return (n*(n-1))/2

[tri_num(n) for n in range(4)]
```

```python
%%time
def prime_list(num):
    primes = []
    sieve = set()
    for i in xrange(2, num+1):
        if i not in sieve:
            primes.append(i)
            sieve.update(range(i, num+1, i))
    return primes

primes = prime_list(200000)
```

```python
def find_prime_factors(num, primelist):
    factors = []
    new_num = num
    
    # grab the prime factors
    for p in primelist:
        if new_num <= 1:
            return factors
        else:
            while new_num != 1 and not new_num % p:
                # this see's how many hits we have
                factors.append(p)
                if new_num % p:
                    break
                else:
                    #and run through this again
                    new_num /= p
                
    else:
        raise Exception("Prime Number List Too Small")
```

```python
%%cython
# cython: cdivision=True

ctypedef Py_ssize_t index_t

cdef inline long tri_num(long n): 
    return (n*(n-1))/2

cpdef long find_tri_div_cnt(int divisors):
    
    cdef:
        int cnt = 0
        index_t i
        long j = 2
        long cur_num
        
    while True:
        j += 1
        cur_num = tri_num(j)
        cnt = 2
        
        for i in xrange(2, (cur_num / 2) + 1):
            if (cur_num % i) == 0:
                cnt += 1
        
        if cnt >= divisors:
            break
            
    return cur_num
        
        
```
